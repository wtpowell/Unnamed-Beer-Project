```python
import pandas as pd
import numpy as np
```

```python
beer_df = pd.read_csv('checkins.csv')
```

```python
beer_df = beer_df.drop(columns=['created_at', 'checkin_comment', 'user_name', 'beer_slug', 'checkin_id', 'toasts_count', 'avg_rating', 'ratings_count', 'checkins_count'])
```

```python
beer_df = beer_df.drop(columns=['user_id', 'source_app', 'venue_id', 'venue_name', 'venue_country', 'venue_city', 'venue_lat', 'venue_lng'])
beer_df.head()
```

```python
def splitting(beer_name):
    try:
        return(beer_name.split('-')[0])
    except:
        return('NA')
beer_df['cleaned_beer_style'] = beer_df.beer_style.apply(splitting)
```
